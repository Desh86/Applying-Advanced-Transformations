```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json

```

```python
hero = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vS1ZstYLwFgwhZnqDsPjtnlHYhJp_cmW55J8JD5mym0seRsaem3px7QBtuFF0LiI7z1PLCkVKAkdO7J/pub?output=csv')
```

```python
hero.head()
```

```python
hero.info()
```

```python
# Make a list of all characters to replace
to_replace = ['|']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    hero['Hero|Publisher'] = hero['Hero|Publisher'].str.replace(char,' ',regex=False)
    
hero['Hero|Publisher'].head()

```

```python
## adding expand=True
hero['Hero|Publisher'].str.split(' ',expand=True)
```

```python
## save the 2 new columns into the dataframe
hero[['Name','Publisher1','Pubisher2','Publisher3','Publisher4']] = hero['Hero|Publisher'].str.split(' ',expand=True)
hero.head(2)
```

```python
## drop the original column 
hero = hero.drop(columns=['Hero|Publisher', 'Pubisher2','Publisher3','Publisher4'])
hero.head(2)

```

```python
## adding expand=True
hero['Measurements'].str.split(' ',expand=True)
```

```python
## save the 2 new columns into the dataframe
hero[['H','Height','cm','w','Weight','kg']] = hero['Measurements'].str.split(' ',expand=True)
hero.head(2)
```

```python
## drop the original column 
hero = hero.drop(columns=['Measurements', 'H','cm','w','kg'])
hero.head(2)
```
